```python
#dataset from Kaggle - https://www.kaggle.com/maxhorowitz/nflplaybyplay2009to2016

#import libraries used
import pandas as pd
import numpy as np

#read in data
nfl_data = pd.read_csv(r'D://Python//Data//NFL_pbp_0917_v4.csv')
```

```python
#look at data- check for NAN and other issues
nfl_data.sample(5)
```

```python
#want to know how many missng data points per col
missing_values_count = nfl_data.isnull().sum()

#check the missing values for the first 10 cols
missing_values_count[:10]
```

```python
#what's that as a % of all data
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

(total_missing / total_cells) *100
```

```python
#drop nas - based on cols with min 1 missing value
columns_with_na_drop = nfl_data.dropna(axis=1)
columns_with_na_drop.head()
```

```python
#how many columns did we lose?
print("Columns in original dataset: %d \n" % nfl_data.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_drop.shape[1])
```

```python
#fill all missing data with 0 as seperate df

filled_nfl_data = nfl_data.fillna(0)
filled_nfl_data.head(20)
```
